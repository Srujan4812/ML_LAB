```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier

# 1. Create a pandas DataFrame for the given custom dataset
data = {
    'Person': ['A', 'B', 'C', 'D', 'E', 'F'],
    'Age': [18, 23, 24, 41, 43, 38],
    'SalaryK': [50, 55, 70, 60, 70, 40],
    'Target': ['N', 'N', 'N', 'Y', 'Y', 'Y']
}
df_custom = pd.DataFrame(data)

print("Custom Dataset:")
print(df_custom)

# 2. Define the features (Age, SalaryK) and the target variable (Target)
X_custom = df_custom[['Age', 'SalaryK']]
y_custom = df_custom['Target']

print("\nFeatures (X_custom):")
print(X_custom)
print("\nTarget (y_custom):")
print(y_custom)
```

```python
test_data_point = pd.DataFrame({'Age': [35], 'SalaryK': [100]})

print("\nTest Data Point (X):")
print(test_data_point)

# 5. Instantiate a KNeighborsClassifier object with n_neighbors=3
knn_custom = KNeighborsClassifier(n_neighbors=3)

# 6. Train the KNN model using the training features and target variable
knn_custom.fit(X_custom, y_custom)

# 7. Use the trained model to predict the 'Target' for the test data point
predicted_target = knn_custom.predict(test_data_point)

# 8. Print the predicted target for the test data point
print(f"\nPredicted Target for (X, 35, 100): {predicted_target[0]}")
```

```python
##IRIS

import pandas as pd
from sklearn.model_selection import train_test_split

# 1. Load the 'iris (1).csv' dataset
df_iris = pd.read_csv('/content/iris (1).csv')

print("Iris Dataset head:")
print(df_iris.head())

# 2. Separate features (X) and target (y)
# Assuming the last column is the target variable (common for iris dataset)
X_iris = df_iris.drop(columns=df_iris.columns[-1])
y_iris = df_iris[df_iris.columns[-1]]

print("\nIris Features head:")
print(X_iris.head())
print("\nIris Target head:")
print(y_iris.head())

# 3. Split the dataset into training and testing sets (80% training, 20% testing)
X_train_iris, X_test_iris, y_train_iris, y_test_iris = train_test_split(
    X_iris, y_iris, test_size=0.2, random_state=42, stratify=y_iris
)

print(f"\nShape of X_train_iris: {X_train_iris.shape}")
print(f"Shape of X_test_iris: {X_test_iris.shape}")
print(f"Shape of y_train_iris: {y_train_iris.shape}")
print(f"Shape of y_test_iris: {y_test_iris.shape}")
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# 4. Instantiate a KNeighborsClassifier object (choosing k=5 as a common starting point)
knn_iris = KNeighborsClassifier(n_neighbors=5)

# 5. Train the KNN model using the training features and target variable
knn_iris.fit(X_train_iris, y_train_iris)

# 6. Make predictions on the test data
y_pred_iris = knn_iris.predict(X_test_iris)

# 7. Calculate and display the accuracy score
accuracy_iris = accuracy_score(y_test_iris, y_pred_iris)
print(f"\nAccuracy Score for Iris dataset: {accuracy_iris:.4f}")

# 8. Generate and display the confusion matrix
conf_matrix_iris = confusion_matrix(y_test_iris, y_pred_iris)
print("\nConfusion Matrix for Iris dataset:")
print(conf_matrix_iris)

# 9. Generate and display the classification report
class_report_iris = classification_report(y_test_iris, y_pred_iris)
print("\nClassification Report for Iris dataset:")
print(class_report_iris)
```

```python
##DIABETES

import pandas as pd

# 1. Load the 'diabetes.csv' dataset
df_diabetes = pd.read_csv('/content/diabetes.csv')

print("Diabetes Dataset head:")
print(df_diabetes.head())
```

```python
import pandas as pd

# 2. Separate features (X) and target (y)
X_diabetes = df_diabetes.drop('Outcome', axis=1)
y_diabetes = df_diabetes['Outcome']

print("\nDiabetes Features head:")
print(X_diabetes.head())
print("\nDiabetes Target head:")
print(y_diabetes.head())
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# 3. Apply feature scaling using StandardScaler to the features
scaler = StandardScaler()
X_diabetes_scaled = scaler.fit_transform(X_diabetes)

print("\nScaled Diabetes Features (first 5 rows):")
print(X_diabetes_scaled[:5])

# 4. Split the dataset into training and testing sets (80% training, 20% testing)
X_train_diabetes, X_test_diabetes, y_train_diabetes, y_test_diabetes = train_test_split(
    X_diabetes_scaled, y_diabetes, test_size=0.2, random_state=42, stratify=y_diabetes
)

print(f"\nShape of X_train_diabetes: {X_train_diabetes.shape}")
print(f"Shape of X_test_diabetes: {X_test_diabetes.shape}")
print(f"Shape of y_train_diabetes: {y_train_diabetes.shape}")
print(f"Shape of y_test_diabetes: {y_test_diabetes.shape}")
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

# 5. Instantiate a KNeighborsClassifier object with a chosen K value (e.g., k=5)
knn_diabetes = KNeighborsClassifier(n_neighbors=5)

# 6. Train the KNN model using the scaled training data
knn_diabetes.fit(X_train_diabetes, y_train_diabetes)

# 7. Make predictions on the scaled test data
y_pred_diabetes = knn_diabetes.predict(X_test_diabetes)

# 8. Calculate and display the accuracy score
accuracy_diabetes = accuracy_score(y_test_diabetes, y_pred_diabetes)
print(f"\nAccuracy Score for Diabetes dataset: {accuracy_diabetes:.4f}")

# 9. Generate and display the confusion matrix
conf_matrix_diabetes = confusion_matrix(y_test_diabetes, y_pred_diabetes)
print("\nConfusion Matrix for Diabetes dataset:")
print(conf_matrix_diabetes)
```

```python
##HEART DISEASE


import pandas as pd

# 1. Load the 'heart.csv' dataset
df_heart = pd.read_csv('/content/heart.csv') # Assuming heart.csv is available in /content/

print("Heart Disease Dataset head:")
print(df_heart.head())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import numpy as np

# 3. Split the dataset into training and testing sets (80% training, 20% testing)
X_train_heart, X_test_heart, y_train_heart, y_test_heart = train_test_split(
    X_heart, y_heart, test_size=0.2, random_state=42, stratify=y_heart
)

print(f"\nShape of X_train_heart: {X_train_heart.shape}")
print(f"Shape of X_test_heart: {X_test_heart.shape}")
print(f"Shape of y_train_heart: {y_train_heart.shape}")
print(f"Shape of y_test_heart: {y_test_heart.shape}")

# 4. Iterate through different n_neighbors (K values) to find the one that yields the highest accuracy score
# We'll try K from 1 to 20
accuracy_scores = []
k_values = range(1, 21)

for k in k_values:
    knn_heart_tuned = KNeighborsClassifier(n_neighbors=k)
    knn_heart_tuned.fit(X_train_heart, y_train_heart)
    y_pred_heart_tuned = knn_heart_tuned.predict(X_test_heart)
    accuracy_scores.append(accuracy_score(y_test_heart, y_pred_heart_tuned))

# Find the optimal K value
optimal_k_heart = k_values[np.argmax(accuracy_scores)]
max_accuracy_heart = np.max(accuracy_scores)

print(f"\nOptimal K value for Heart dataset: {optimal_k_heart}")
print(f"Maximum Accuracy for Heart dataset with optimal K: {max_accuracy_heart:.4f}")

# Plotting accuracy vs K value
plt.figure(figsize=(10, 6))
plt.plot(k_values, accuracy_scores, marker='o')
plt.title('Accuracy vs. K Value for Heart Disease Dataset')
plt.xlabel('Number of Neighbors (K)')
plt.ylabel('Accuracy')
plt.xticks(k_values)
plt.grid(True)
plt.show()

# 5. Build a KNN classifier with the optimal K value
knn_heart_optimal = KNeighborsClassifier(n_neighbors=optimal_k_heart)

# 6. Train the KNN model using the training data
knn_heart_optimal.fit(X_train_heart, y_train_heart)

# 7. Make predictions on the test data
y_pred_heart_optimal = knn_heart_optimal.predict(X_test_heart)

# 8. Calculate and display the accuracy score of the trained model on the test data
accuracy_optimal_heart = accuracy_score(y_test_heart, y_pred_heart_optimal)
print(f"\nAccuracy Score for Heart dataset with optimal K ({optimal_k_heart}): {accuracy_optimal_heart:.4f}")

# 9. Generate and display the confusion matrix
conf_matrix_optimal_heart = confusion_matrix(y_test_heart, y_pred_heart_optimal)
print("\nConfusion Matrix for Heart dataset with optimal K:")
print(conf_matrix_optimal_heart)

# 10. Generate and display the classification report
class_report_optimal_heart = classification_report(y_test_heart, y_pred_heart_optimal)
print("\nClassification Report for Heart dataset with optimal K:")
print(class_report_optimal_heart)
```

```python
import pandas as pd
import io
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Convert the class_report_optimal_heart string into a pandas DataFrame
report_data = io.StringIO(class_report_optimal_heart)

# Read the classification report, explicitly defining the separator, skipping initial spaces,
# and setting the first column (containing class labels/summary rows) as the index.
# Using a raw string for the regex separator to avoid SyntaxWarning.
df_report = pd.read_csv(report_data, sep=r'\s{2,}', engine='python', skipinitialspace=True, index_col=0)

# Set the index name to 'Classes'
df_report.index.name = 'Classes'

# 2. Clean the DataFrame: remove rows that contain 'accuracy', 'macro avg', and 'weighted avg'
df_report = df_report.drop(index=['accuracy', 'macro avg', 'weighted avg'], errors='ignore')

# Remove columns that are not relevant for plotting, keep 'precision', 'recall', 'f1-score'
df_report = df_report[['precision', 'recall', 'f1-score']]

# Convert to numeric, errors='coerce' will turn non-numeric into NaN
df_report = df_report.apply(pd.to_numeric, errors='coerce')

# 3. Create a heatmap of the processed classification report DataFrame
plt.figure(figsize=(8, 6))
sns.heatmap(df_report, annot=True, cmap='Blues', fmt=".2f")

# 4. Label the x-axis with 'Metrics' and the y-axis with 'Classes'
plt.xlabel('Metrics')
plt.ylabel('Classes')

# 5. Add a title to the plot
plt.title('Classification Report for Heart Disease Prediction (Optimal K)')

# 6. Display the plot
plt.show()

print("Classification Report DataFrame:")
print(df_report)
```
